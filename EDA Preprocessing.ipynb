## OUTLIERS FORMULA
  ## IQR-METHOD

```python
for col in df.select_dtypes(include="number").columns:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    print(IQR)
    print(f"IQR for {col}: {IQR}")
```

```python
df=df[~((df[col]<(Q1-1.5*IQR))|(df[col]>(Q3+1.5*IQR)))]
df.shape
```

## REMOVING OUTLIERS AND SHOW THE BOXPLOT

```python
for i in df.select_dtypes(include="number").columns:
    sns.boxplot(df[i])
    plt.show()
```

## BARPLOT 
 ## YEARLY SALES OF TRANSMISSION TYPES

```python

col_count = df.groupby(['Year', 'Transmission']).size().reset_index(name='Count')
top_transmissions = col_count.groupby('Transmission')['Count'].sum().nlargest(20).index
filtered_col_count = col_count[col_count['Transmission'].isin(top_transmissions)]

plt.figure(figsize=(10,5))
sns.barplot(data=filtered_col_count, x='Year', y='Count', hue='Transmission')
plt.title("Yearly Sales of Transmission Types")
plt.ylabel("Number of Sales")
plt.xlabel("Year")
plt.legend(title='Transmission Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)  
plt.show()
```
